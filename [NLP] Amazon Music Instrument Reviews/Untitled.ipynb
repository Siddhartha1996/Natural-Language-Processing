# Preprocessing the actual text data

```python
df1["text"] = df1["text"].str.lower()
df1.head()
```

```python
import string

punctuation = string.punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', punctuation))

df1["text"] = df1["text"].apply(lambda text: remove_punctuation(text))
df1.head()
```

```python
stopwords = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in stopwords])

df1["text"] = df1["text"].apply(lambda text: remove_stopwords(text))
df1.head()
```

```python
# Frequent words removal
from collections import Counter
cnt = Counter()

FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

df1["text"] = df1["text"].apply(lambda text: remove_freqwords(text))
df1.head()
```

```python
#Removal of rare words:

RAREWORDS = set([w for (w, wc) in cnt.most_common()[:-10-1:-1]])
def remove_rarewords(text):
    """custom function to remove the rare words"""
    return " ".join([word for word in str(text).split() if word not in RAREWORDS])

df1["text"] = df1["text"].apply(lambda text: remove_rarewords(text))
df1.head()
```

```python
import nltk
# nltk.download('averaged_perceptron_tagger')
# Lemmatizing the data:

from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
wordnet_map = {"N":wordnet.NOUN, "V":wordnet.VERB, "J":wordnet.ADJ, "R":wordnet.ADV}
def lemmatize_words(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text])

df1["text_lemmatized"] = df1["text"].apply(lambda text: lemmatize_words(text))
df1.head()
```

```python
df1 = df1.drop(['text'], axis=1)
```

```python
df1 = df1.rename(columns={"text_lemmatized": "text"})
df1.head()
```

# Vectorizing the text

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
vectorizer.fit(df1['text'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1['text'], df1['overall'], test_size=0.25, random_state=21)
```

```python
X_train.shape, X_test.shape
```

```python
X_train_tfidf = vectorizer.transform(X_train)

X_test_tfidf = vectorizer.transform(X_test)
```

```python
# K Nearest Neighbour

from sklearn import neighbors 
from sklearn.metrics import f1_score,confusion_matrix,roc_auc_score
f1_list=[]
k_list=[]
for k in range(1,10):
    clf=neighbors.KNeighborsClassifier(n_neighbors=k,n_jobs=-1)
    clf.fit(X_train_tfidf,y_train)
    pred=clf.predict(X_test_tfidf)
    f=f1_score(y_test,pred,average='macro')
    f1_list.append(f)
    k_list.append(k)
```

```python
best_f1_score=max(f1_list)
best_k=k_list[f1_list.index(best_f1_score)]        
print("Optimum K value=",best_k," with F1-Score=",best_f1_score)
```

```python
KNN_model=neighbors.KNeighborsClassifier(n_neighbors=4,n_jobs=-1)
KNN_model.fit(X_train_tfidf,y_train)
pred=KNN_model.predict(X_test_tfidf)
print("Accuracy={}%".format((sum(y_test==pred)/y_test.shape[0])*100))
```
